### ¿Por qué usar NumPy? 

Las listas de Python son excelentes contenedores de propósito general. Pueden ser "heterogéneas", lo que significa que pueden contener elementos de diversos tipos, y son bastante rápidas cuando se utilizan para realizar operaciones individuales con un puñado de elementos.

Dependiendo de las características de los datos y los tipos de operaciones que se deban realizar, otros contenedores pueden ser más apropiados. Al aprovechar estas características, podemos mejorar la velocidad, reducir el consumo de memoria y ofrecer una sintaxis de alto nivel para realizar diversas tareas de procesamiento comunes. NumPy destaca cuando se procesan grandes cantidades de datos homogéneos (del mismo tipo) en la CPU.

El objeto principal de NumPy es el array multidimensional homogéneo. Es una tabla de elementos (generalmente números), todos del mismo tipo, indexados por una tupla de enteros no negativos. En NumPy, las dimensiones se denominan ejes .

Por ejemplo, el array de coordenadas de un punto en el espacio 3D, tiene un eje. Este eje contiene 3 elementos, por lo que decimos que tiene una longitud de 3. En el ejemplo que se muestra a continuación, el array tiene 2 ejes. El primer eje tiene una longitud de 2 y el segundo, de 3.[1, 2, 1]

```python
[[1., 0., 0.],
 [0., 1., 2.]]
```

La clase de array de NumPy se llama ndarray. También se conoce con el alias array. Tenga en cuenta que numpy.arrayno es lo mismo que la clase , de la biblioteca estándar de Python array.array, que solo maneja arrays unidimensionales y ofrece menos funcionalidad. Los atributos más importantes de un ndarrayobjeto son:

**ndarray.ndim**
el número de ejes (dimensiones) de la matriz.

**ndarray.forma**
Las dimensiones del array. Esta es una tupla de enteros que indica el tamaño del array en cada dimensión. Para una matriz con n filas y m columnas, shapeserá (n,m). Por lo tanto, la longitud de la shapetupla es el número de ejes, ndim.

**ndarray.size**
El número total de elementos del array. Es igual al producto de los elementos de shape.

**ndarray.dtype**
Un objeto que describe el tipo de los elementos del array. Se pueden crear o especificar tipos de datos (dtypes) utilizando tipos estándar de Python. Además, NumPy proporciona sus propios tipos. numpy.int32, numpy.int16 y numpy.float64 son algunos ejemplos.

**ndarray.tamaño del artículo**
El tamaño en bytes de cada elemento del array. Por ejemplo, un array de elementos de tipo float64tiene itemsize8 (=64/8), mientras que uno de tipo complex32tiene itemsize4 (=32/8). Equivale a ndarray.dtype.itemsize.

**ndarray.data**
El búfer que contiene los elementos del array. Normalmente, no necesitamos usar este atributo, ya que accederemos a los elementos del array mediante herramientas de indexación.

### Principales diferencias con las listas habituales de Python

El primer punto es diferenciar los arrays de NumPy de las listas de Python, ya que no se comportan de la misma manera. Definamos dos listas de Python y los dos arrays equivalentes de NumPy.

```python
import numpy as np
l1, l2 = [1, 2, 3], [3, 4, 5]
a1, a2 = np.array([1, 2, 3]), np.array([3, 4, 5])
print(l1, l2)
```

En primer lugar, todas las operaciones matemáticas actúan elemento por elemento en un array de NumPy. Para las listas de Python, la suma actúa como una concatenación de las listas, y una multiplicación por un escalar actúa como una replicación de las listas:


```python
# obj1 + obj2
print('listas de Python: {}'.format(l1 + l2))
print('arrays de NumPy: {}'.format(a1 + a2))
```

```python
# obj * 3
print('lista de Python: {}'.format(l1 * 3))
print('array de NumPy: {}'.format(a1 * 3))
```

Otra diferencia importante es la forma de acceder a los elementos de un array, el llamado *slicing* e indexación. Aquí, el comportamiento de las listas de Python y los arrays de NumPy es más similar, excepto que los arrays de NumPy admiten algunas características más, como la indexación mediante un array de enteros (que no funciona para las listas de Python). Los casos de uso de dicha indexación se ilustrarán ampliamente en los próximos capítulos.


```python
# Indexación con un entero: obj[1]
print('lista de Python: {}'.format(l1[1]))
print('array de NumPy: {}'.format(a1[1]))
```

```python
# Indexación con un slicing: obj[slice(1,3)]
print('lista de Python: {}'.format(l1[slice(1, 3)]))
print('array de NumPy: {}'.format(l1[slice(1, 3)]))
```

```python
# Indexación con una lista de enteros: obj[[0,2]]
print('lista de Python: IMPOSIBLE')
print('array de NumPy: {}'.format(a1[[0, 2]]))
```

### Principales características de un array
La fuerza de los arrays de NumPy es que son multidimensionales. Esto permite describir un conjunto de datos complejo en un solo array de NumPy, sobre el cual se pueden realizar operaciones. En NumPy, las dimensiones también se llaman *ejes*. Por ejemplo, un conjunto de 2 posiciones en el espacio $\\vec{r}_i$ puede verse como un array 2D de NumPy, con el primer eje siendo el punto $i=1$ o $i=2$, y el segundo eje siendo las coordenadas ($x,y,z$). Hay algunos atributos que describen arrays multidimensionales:

- `a.dtype`: tipo de datos contenidos en el array.
- `a.shape`: número de elementos a lo largo de cada dimensión (o eje).
- `a.size`: número total de elementos (producto de los elementos de `a.shape`).
- `a.ndim`: número de dimensiones (o ejes).

```python
points = np.array([[0, 1, 2],
                   [3, 4, 5]])

print('a.dtype = {}'.format(points.dtype))
print('a.shape = {}'.format(points.shape))
print('a.size = {}'.format(points.size))
print('a.ndim = {}'.format(points.ndim))
```

## Las tres características clave de NumPy

### Vectorización

La *vectorización* es una forma de realizar cálculos en arrays de NumPy **sin bucles explícitos**, que son muy lentos en Python. La idea de la vectorización es calcular una operación dada *elemento por elemento* mientras la operación se llama al array mismo. A continuación se da un ejemplo para calcular el inverso de 100000 números, tanto con un bucle explícito como con vectorización.


```python
a = np.random.randint(low=1, high=100, size=100000)

def explicit_loop_for_inverse(array):
    res = []
    for a in array:
        res.append(1./a)
    return np.array(res)
# Usando un bucle explícito
%timeit explicit_loop_for_inverse(a)
```

```python
# Usando comprensión de listas
%timeit [1./x for x in a]
```

```python
# Usando vectorización
%timeit 1./a
```

**La supresión de bucles *for* explícitos es probablemente el aspecto más desconocido de NumPy, según mi opinión, y merece un poco de práctica. Al final, las líneas de código se vuelven relativamente cortas, pero hay que pensar bien cómo implementar un cálculo dado de una manera *pythónica*.**

Muchas funciones estándar están implementadas de manera vectorizada, se llaman *funciones universales*, o `ufunc`. A continuación se dan algunos ejemplos, pero la descripción completa se puede encontrar en la [documentación de NumPy](https://docs.scipy.org/doc/numpy-1.15.1/reference/ufuncs.html).


```python
a = np.random.randint(low=1, high=100, size=3)
print('a        : {}'.format(a))
print('a^2      : {}'.format(a**2))
print('a/(1-a^a): {}'.format(a/(1-a**a)))
print('cos(a)   : {}'.format(np.cos(a)))
print('exp(a)   : {}'.format(np.exp(a)))
```

Todas estas funciones universales pueden funcionar para arrays de n dimensiones y se pueden usar de manera muy flexible dependiendo del eje al que se refieran. De hecho, la operación matemática puede realizarse sobre un eje diferente del array, teniendo un significado totalmente diferente. Daremos un ejemplo simple concreto con un array 2D de forma (5,2), es decir, 5 vectores de tres coordenadas $(x,y,z)$. Se discutirán muchos más ejemplos en la sección 2.


```python
# Generar 5 vectores (x,y,z)
positions = np.random.randint(low=1, high=100, size=(5, 3))

# Promedio de las coordenadas sobre las 5 observaciones
pos_mean = np.mean(positions, axis=0)
print('mean = {}'.format(pos_mean))

# Distancia al origen sqrt(x^2 + y^2 + z^2) para las 5 observaciones
distances = np.sqrt(np.sum(positions**2, axis=1))
print('distances = {}'.format(distances))
```

### Broadcasting

El *broadcasting* es una forma de calcular operaciones entre arrays de diferentes tamaños de manera implícita (y concisa). Un ejemplo concreto podría ser trasladar tres posiciones $\\vec{r}_i=(x,y)_i$ por un vector $\\vec{d}_0$ simplemente sumando `points + d0` donde `points.shape=(3,2)` y `d0.shape=(2,)`. A continuación se dan algunos ejemplos, pero se pueden encontrar más detalles en [esta documentación](https://docs.scipy.org/doc/numpy-1.13.0/user/basics.broadcasting.html).


```python
# Operación entre formas (3) y (1)
a = np.array([1, 2, 3])
b = np.array([5])
print('a + b = \n{}'.format(a + b))
```

```python
# Operación entre formas (3) y (1,2)
a = np.array([1, 2, 3])
b = np.array([
              [4],
              [5],
             ])
print('a + b = \n{}'.format(a + b))
```

```python
# Trasladar 3 vectores 2D por d0=(1,4)
points = np.random.normal(size=(3, 2))
d0 = np.array([1, 4])
print('points:\n {}\\n'.format(points))
print('points + d0:\n {}'.format(points + d0))
```

No todas las formas pueden combinarse, y hay *reglas de broadcasting*, que son (citando la [documentación de NumPy](https://docs.scipy.org/doc/numpy-1.15.0/user/basics.broadcasting.html)):

> Cuando se opera en dos arrays, NumPy compara sus formas elemento por elemento. Comienza con las dimensiones finales y avanza hacia adelante. Dos dimensiones son compatibles cuando:
> 
> 1. Son iguales, o
> 2. Una de ellas es 1.
  
En un caso fallido, se puede agregar un nuevo *eje vacío* `np.newaxis` a un array para igualar su dimensión y hacer posible el broadcasting. Aquí hay un ejemplo muy simple:


```python
a = np.arange(10).reshape(2,5)
b = np.array([10,20])
```

```python
try:
    res = a + b
    print('Posible para {} y {}:'.format(a.shape, b.shape))
    print('a + b = \\n {}'.format(res))
except ValueError:
    print('Imposible para {} y {}'.format(a.shape, b.shape))
```

```python
c = b[:, np.newaxis]
try: 
    res = a + c
    print('Posible para {} y {}:'.format(a.shape, c.shape))
    print('a + c = \\n {}'.format(res))
except ValueError:
    print('Broadcasting para {} y {}'.format(a.shape, c.shape))
```

### Trabajar con subarrays: slicing, indexación y máscaras (o selección)

Como se mencionó anteriormente, el *slicing y la indexación* son formas de acceder a elementos o subarrays de manera inteligente. Python permite el slicing con el objeto `Slice()`, pero `numpy` permite llevar la lógica mucho más allá con lo que se llama *fancy indexing*. A continuación se dan algunos ejemplos, y para más detalles, por favor echa un vistazo a [esta página de documentación](https://docs.scipy.org/doc/numpy-1.15.1/reference/arrays.indexing.html).

**Regla 1:** La sintaxis es `a[i]` para acceder al i-ésimo elemento. También es posible ir desde el último elemento usando índices negativos: `a[-1]` es el último elemento.


```python
a = np.random.randint(low=1, high=100, size=10)
print('a = {}'.format(a))
print('a[2] = {}'.format(a[2]))
print('a[-1] = {}'.format(a[-1]))
print('a[[1, 2, 5]] = {}'.format(a[[1, 2, 5]]))
```

**Regla 2:** NumPy también admite arrays de índices. Si el array de índices es multidimensional, el array devuelto tendrá la misma dimensión que el array de índices.


```python
# Pequeño array de índices n-dimensional: 3 arrays de 2 elementos
indices = np.arange(6).reshape(3,2)
print('indices =\n {}'.format(indices))
print('a[indices] =\n {}'.format(a[indices]))
```

### Arrays de índices n-dimensionales: 2 arrays de (10, 10) arrays

```python
indices_big = np.random.randint(low=0, high=10, size=(2, 3, 2))
print('indices_big =\n {}'.format(indices_big))
print('a[indices_big] =\n {}'.format(a[indices_big]))
```

**Regla 3:** Hay una forma inteligente de acceder a subarrays con la sintaxis `a[min:max:step]`. De esta manera, por ejemplo, es muy fácil tomar un elemento de cada dos (`step=2`), o invertir el orden de un array (`step=-1`). Esta sintaxis también funciona para arrays n-dimensionales, donde cada dimensión está separada por una coma. Se da un ejemplo para un array 1D y para un array 3D de forma (5, 2, 3), que puede considerarse como 5 observaciones de 2 posiciones en el espacio.


```python
# Array 1D
a = np.random.randint(low=1, high=100, size=10)
print('array completo a              = {}'.format(a))
print('de 0 a 1: a[:2]        = {}'.format(a[:2]))
print('de 4 al final: a[4:]      = {}'.format(a[4:]))
print('orden inverso: a[::-1]    = {}'.format(a[::-1]))
print('todos los elementos pares: a[::2] = {}'.format(a[::2]))
```

```python
# Array 3D
a = np.random.randint(low=0, high=100, size=(5, 2, 3))
print('a = \n{}'.format(a))
```

Digamos que uno quiere tomar solo las coordenadas $(x,y)$ para el primer vector para las 5 observaciones. Así es como se dividirá cada eje:
- Primer eje (=5 observaciones): `:`, es decir, tomar todo.
- Segundo eje (=2 vectores): `1`, es decir, solo el segundo elemento.
- Tercer eje (=3 coordenadas): `0:2`, es decir, desde $0$ hasta $2-1=1$, así que solo $(x,y)$.

```python
# Tomando solo los valores x,y del primer vector para todas las observaciones:
print('a[:, 0, 0:2] =\n {}'.format(a[:, 0, 0:2]))
```

```python
# Invertir el orden de los 2 vectores para cada observación:
print('a[:, ::-1, :] = \n{}'.format(a[:, ::-1, :]))
```

**Regla 4:** La última parte de la indexación es sobre *enmascarar* arrays o, en un lenguaje más común, *seleccionar* subarrays/elementos. Esto permite obtener solo elementos que satisfacen un criterio dado, explotando las reglas de indexación descritas anteriormente. De hecho, una operación booleana aplicada a un array como `a > 0` devolverá directamente un array de valores booleanos `True` o `False` dependiendo de si el elemento correspondiente satisface la condición o no.


```python
a = np.random.randint(low=-100, high=100, size=(5, 3))
mask = a > 0
print('a = \n{}'.format(a))
print('\nmask = \n {}'.format(mask))
```

```python
print('\na[mask] = \n {}'.format(a[mask])) # siempre devuelve un array 1D
print('\na * mask = \n {}'.format(a * mask)) # preserva la dimensión (False=0)
print('\na[~mask] = \n {}'.format(a[~mask])) # ~mask es la negación de mask
print('\na * ~mask  = \n {}'.format(a * ~mask)) # también funciona para un producto.
```


**Nota:** El caso de arrays booleanos como índices tiene un tratamiento especial en NumPy (ya que el resultado siempre es un array 1D). En realidad, hay un objeto dedicado de NumPy llamado *array enmascarado* (cf. [documentación](https://docs.scipy.org/doc/numpy-1.15.1/reference/maskedarray.html)) que permite mantener todo el array pero sin considerar algunos elementos en el cálculo (por ejemplo, una cámara CCD con píxeles muertos). Sin embargo, cuando un array booleano se usa en una operación matemática (como `a * mask`), entonces `False` se trata como `0` y `True` como `1`:


```python
print('a + mask = \n{}'.format(a + mask))
```


Estos arrays booleanos también son muy útiles para *reemplazar una categoría de elementos* con un valor dado de una manera muy fácil, concisa y legible:


```python
a = np.random.randint(low=-100, high=100, size=(5, 3))
print('Antes: a=\n{}'.format(a))

a[a < 0] = a[a < 0]**2
print('\nDespués: a=\n{}'.format(a))
```

#### Ejemplo de gráficos 1D e histogramas

Para jugar con los datos, generamos 2 muestras de 1000 valores distribuidos según una función de densidad de probabilidad normal con $\\mu=-1$ y $\\mu=1$ respectivamente, y $\\sigma=0.5$. Estos datos se almacenan en un array de NumPy `x` de forma `x.shape=(1000, 2)`. Luego simplemente calculamos y almacenamos el seno de todos estos valores en un array de la misma forma `y`:


```python
x = np.random.normal(loc=[-1, 1], scale=[0.5, 0.5], size=(1000,2))
y = np.sin(x)
```

El siguiente paso es trazar estos datos de dos maneras: primero queremos `y` vs. `x`, segundo queremos el histograma de los valores de `x`. Necesitamos primero crear una figura, luego crear dos *subplots* (especificando el número de líneas, columnas e índice de subplot). Tenga en cuenta que matplotlib siempre toma la primera dimensión para definir los números a graficar, mientras que las dimensiones superiores se consideran otros gráficos, superpuestos automáticamente.


```python
plt.figure(figsize=(24, 10))
plt.subplot(121) # 121 significa 1 línea, 2 columnas, primer gráfico
plt.plot(x, y, marker='o', markersize=5, linewidth=0.0)
plt.subplot(122) # 122 significa 1 línea, 2 columnas, segundo gráfico
plt.hist(x, bins=20);
```

### Ejemplo de diagrama de dispersión 2D

Un diagrama de dispersión permite dibujar un marcador en un espacio 2D y una tercera información se codifica en el tamaño del marcador. Para jugar, generamos dos conjuntos de 5000 números distribuidos según gaussianas no correlacionadas de $(\mu_0=\mu_1=0)$ y $(\sigma_1,\sigma_2)=(0.5, 0.8)$ en un array numpy `points` con forma `points.shape=(5000,2)`. Estos dos conjuntos de números se interpretan como posiciones $(x,y)$ que se cargan en dos arrays `(5000, 1)` `x` e `y`:

```python
points = np.random.normal(loc=[0, 0], scale=[0.5, 0.8], size=(5000,2))
x, y = points[:, 0], points[:, 1]
```

Luego podemos trazar los 5000 puntos en el plano 2D, y aquí especificamos el tamaño del marcador en $100\times\sin^2(x)$ usando el argumento `s` de la función `plt.scatter()` (tenga en cuenta que la matriz `x`, `y` y `s` deben tener la misma forma):

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, s=100*(np.sin(x))**2, marker='o', alpha=0.3)
plt.xlim(-3, 3)
plt.ylim(-3, 3);
```

#### Ejemplo de gráficos 3D

Para gráficos 3D, se pueden generar 1000 posiciones en el espacio y operar una traslación mediante un vector $\vec{r}_0$ mediante difusión:

```python
data = np.random.normal(size=(1000, 3))
r0 = np.array([1, 4, 2])
data_trans = data + r0
```

Luego es fácil recuperar las coordenadas espaciales iniciales (es decir, antes de la traducción) y finales (es decir, después de la traducción):

```python
xi, yi, zi = data[:,0], data[:,1], data[:,2]
xf, yf, zf = data_trans[:,0], data_trans[:,1], data_trans[:,2]
```

Se debe importar un módulo adicional para graficar datos en tres dimensiones, y se debe definir la proyección. Una vez hecho esto, una simple llamada a `ax.scatter3D(x,y,z)` realiza el gráfico. Tenga en cuenta que llamamos a una función de `ax` y no a `plt` como antes. Esto se debe al comando `ax = plt.axes(projection='3d')`, necesario para graficar en 3D. Para más detalles, consulte el [tutorial 3D de matplotlib](https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html).

```python
from mpl_toolkits import mplot3d
plt.figure(figsize=(12,10))
ax = plt.axes(projection='3d')
ax.scatter3D(xi, yi, zi, alpha=0.4, label='Antes ')
ax.scatter3D(xf, yf, zf, alpha=0.4, label='Despues')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.legend(frameon=False, fontsize=18);
```

```python
from scipy import optimize
from scipy import stats
```

Ahora, mostremos cómo realizar un ajuste de datos con barra de error usando una función específica de `scipy.optimize`. Primero, necesitamos generar datos con 20 mediciones seleccionadas, con un ruido de aproximadamente el 30 % y una incertidumbre combinada de 0,1 % de incertidumbre absoluta y 10 % de incertidumbre relativa.

```python
Npoints, Nsampling = 20, 1000
xcont = np.linspace(-5.0, 3.5, Nsampling)
x = np.linspace(-5, 3.0, Npoints)
y = 2*(np.sin(x/2)**2 + np.random.random(Npoints)*0.3)
dy = np.sqrt(0.10**2 + (0.10*y)**2)
```

Luego, necesitamos definir las funciones con las que queremos ajustar nuestros datos, por ejemplo, polinomios de grado 1. La sintaxis debe ser `func(x, *pars)`:

```python
def pol1(x, p0, p1):
    return p0 + x*p1
```

Las siguientes líneas realmente realizan el ajuste y devuelven tanto los parámetros óptimos como las covarianzas para el polinomio de grado 1:

```python
p, cov = optimize.curve_fit(pol1, x, y, sigma=dy)
```

Se puede generalizar el procedimiento graficando el resultado del ajuste para polinomios de varios grados, después de graficar los datos. Esta es una buena manera de comparar diferentes modelos para los mismos datos. Primero, definimos un polinomio de grado arbitrario `plo_func()` y lo *vectorizamos* usando `np.vectorize` para que acepte arrays de NumPy:

```python
def pol_func(x, *coeff):
    a = np.array([coeff[i]*x**i for i in range(len(coeff))])
    return np.sum(a)

pol_func = np.vectorize(pol_func)
```

En la llamada anterior a `optimize.curve_fit()`, no usamos argumentos adicionales. Para este ejemplo, necesitamos especificar al menos el punto inicial de los parámetros `p0`, ya que el número de parámetros se evaluará mediante `len(p0)` (no se conoce a priori, ya que se asigna dinámicamente). Se pueden especificar otras opciones, como los valores mínimo y máximo permitidos para los parámetros. A continuación, se muestra una función de ajuste que realiza el ajuste para un grado de polinomio arbitrario:

```python
def fit_polynom(degree):
    nPars = degree+1
    p0, pmin, pmax = [1.0]*nPars, [-10]*nPars, [10]*nPars
    fit_options = {'p0': p0, 'bounds': (pmin, pmax), 'check_finite': True}
    par, cov = optimize.curve_fit(pol_func, x, y, sigma=dy, **fit_options)
    return par, cov

degree_max = 12
```

El siguiente código prueba todas las funciones polinomiales hasta un grado `degree_max=`, realiza el ajuste y superpone el resultado de cada uno junto con los datos experimentales en la misma figura:

```python
# Figura para el resultado
fig = plt.figure(figsize=(12,7))

# Ajuste y trazado
for d in np.arange(0, degree_max):
    par, cov = fit_polynom(d)
    plt.plot(xcont, pol_func(xcont, *par), label='pol{}'.format(d), 
             linewidth=3, alpha=0.8)

# Plot de datos
style = {'marker': 'o', 'color': 'black', 'markersize': 8,
         'linestyle': '', 'zorder': 10, 'label': 'Data'}
plt.errorbar(x, y, yerr=dy, **style)

# Plot adicionales
plt.xlim(-5.5, 6.3)
plt.ylim(-0.3, 3.4)
plt.legend(frameon=False, fontsize='xx-large');
```

Es posible cuantificar la precisión con la que un modelo dado explica las observaciones, calculando lo que llamamos *bondad de ajuste*. En un enfoque frecuentista, esto se puede evaluar mediante la fracción de pseudodatos provenientes, en principio, de repetir exactamente el mismo experimento, con la peor concordancia para un modelo dado. La concordancia se puede cuantificar utilizando $\chi^2 = \sum^{n}_{i=1} \frac{(y_i - f(x_i))^2}{{\sigma_i}^2}$ y su función de densidad de probabilidad (PDF) proporciona acceso directo a la fracción de "peores pseudodatos" (integrando la PDF de $\chi^2$ a $\infty$). Más precisamente, se puede utilizar la función de distribución acumulada (CDF) de $\chi^2$ calculada con $n$ grados de libertad, por ejemplo, `Npoins`, *es decir,* `len(x)`. Se pueden encontrar más detalles, por ejemplo, en la [revisión estadística del Grupo de Datos de Partículas](http://pdg.lbl.gov/2018/reviews/rpp2018-rev-statistics.pdf). Las dos funciones siguientes permiten calcular la bondad del ajuste:

```python
def get_chi2_nDOF(y, dy, yfit):
    r = (y-yfit)/dy
    return  np.sum(r**2), len(y)

def get_pvalue(chi2, nDOF):
    return 1-stats.chi2.cdf(chi2, df=nDOF)
```

Ahora podemos realizar todos estos ajustes y extraer la bondad del ajuste ($\chi^2$ y $p$-valor) para cada modelo:

```python
# Ajuste y obtención del valor p
degree, chiSquare, pvalue = [], [], []
for d in np.arange(degree_max):
    par, cov = fit_polynom(d)
    c2, n = get_chi2_nDOF(y, dy, pol_func(x, *par))
    degree.append(d), chiSquare.append(c2), pvalue.append(get_pvalue(c2, n))
```

El siguiente fragmento de código representa gráficamente el valor $\chi^2$ y el valor $p$ en función del grado del polinomio utilizando dos ejes $y$ diferentes. Esto ofrece otra forma de usar matplolib: definir objetos explícitos como `ax` y `fig` e invocar métodos en ellos (denominado *enfoque sin estado*), en lugar de usar una función en `plt` (denominado *enfoque con estado*). Para más detalles sobre estos diferentes enfoques, consulte esta [publicación de RealPython](https://realpython.com/python-matplotlib-guide/#the-matplotlib-object-hierarchy).

```python
# Trazando el resultado con 2 ejes diferentes
fig, ax1 = plt.subplots(figsize=(12,7))
ax1.set_xlabel('Grado del polinomio', fontsize=20)
style = {'marker': 'o', 'markersize': 10, 'alpha': 0.8, 
         'linestyle': '--', 'linewidth': 3}

# Trazar chi2/n
ax1.semilogy(degree, np.array(chiSquare)/Npoints, color='tab:red', **style)
ax1.set_ylim(0.2, 100)
ax1.set_ylabel('$\chi^2/n$ valor [log scale]', color='tab:red', fontsize=20)

# Graficar valores p
ax2 = ax1.twinx()
ax2.plot(degree, pvalue, color='tab:blue', **style)
ax2.set_ylim(-0.1, 1.1)
ax2.set_ylabel('$\chi^2$ probabilidad [lin scale]', color='tab:blue', fontsize=20);
```

También es posible determinar si dos modelos ofrecen una descripción similar o si uno es mejor que el otro. Esto se denomina prueba F y se basa en la suma de residuos al cuadrado $RSS \equiv \sum (y_i-f(x_i))^2$ (RSS). Para un ajuste de $n$ puntos de datos con un modelo M1 definido por $p_1$ parámetros con $RSS_1$ y un modelo M2 definido por $p_2$ parámetros con $RSS_2$ *con $p_2>p_1$*, la prueba F se define como:

$$
F(M1, M2) = \frac{ \left(\frac{RSS_1 - RSS_2 }{p_2-p_1}\right)} {\left(\frac{RSS_2}{n-p_2}\right)}
$$

Si el valor $p$ de $F$ es cercano a 1,0, significa que ambos modelos son igualmente compatibles con los datos y no hay indicios de que se deba elegir uno. En ese caso, se preferiría el modelo más simple con menos parámetros. Si el valor $p$ de $F$ es cercano a 0, entonces los dos modelos comparados son realmente diferentes y se debe seleccionar uno.

Primero definamos la función RSS y calculémosla sobre todos los polinomios:

```python
def get_RSS(y, yfit):
    return np.sum((y-yfit)**2)

RSS = []
for d in np.arange(degree_max):
    par, cov = fit_polynom(d)
    RSS.append(get_RSS(y=y, yfit=pol_func(x, *par)))
```

Luego, se puede definir una función para comparar dos grados donde $d_2$ debe ser mayor que $d_1$:

```python
def compare_pol_pq(d1, d2):
    
  
    if d1 not in degree or d2 not in degree:
        raise NameError('Degree no supported')
    if d2<d1:
        d2, d1 = d1, d2
    
   # Extraer los números M1 y M2
    RSS1, p1 = RSS[d1], d1+1
    RSS2, p2 = RSS[d2], d2+1
    
    # F-value & p-value
    Fval = (RSS1-RSS2)/(p2-p1) * (Npoints-p2)/RSS2
    pval = 1-stats.f.cdf(x=Fval, dfn=p2-p1, dfd=Npoints-p2)
    
    return Fval, pval
```

Grafique los valores de la prueba F y sus valores p para la comparación entre $d-1$ y $d$ a partir de $d=3$. Este gráfico muestra que, a partir de $d=6$, existe una compatibilidad aceptable entre los modelos (en el peor de los casos, del 15%).

```python
# Cálculo de comparaciones de la prueba F a partir de grado=4
dmin, compare = 3, compare_pol_pq
dinf, dsup = np.arange(dmin-1, degree_max), np.arange(dmin, degree_max)
result = np.array([compare(d1, d2) for d1, d2 in zip(dinf, dsup)])
Ftest, pval = result[:, 0], result[:, 1]

# Representación gráfica de la figura y estilo de los resultados
fig, ax1 = plt.subplots(figsize=(12,7))
ax1.set_xlabel('Grado de polinomio', fontsize=20)
style = {'marker': 'o', 'markersize': 10, 'alpha': 0.8,
'linestyle': '--', 'linewidth': 3}

# Gráfico de valores de la prueba F
ax1.plot(dsup, Ftest, color='tab:red', **style)
ax1.set_ylabel('F-test $F(d-1,d)$', color='tab:red', fontsize=20)

# Gráfico de valores p
ax2 = ax1.twinx()
ax2.plot(dsup, pval, color='tab:blue', **style)
ax2.set_ylabel('F-test probabilidad', color='tab:blue', fontsize=20);
```

Para verificar visualmente de manera efectiva que la prueba F brinda información sensata, se puede graficar la suma acumulada del residuo de ajuste $(y-f(x))^2$ y la probabilidad de la prueba F una al lado de la otra:

```python
# Representación gráfica de la figura y estilo de los resultados
plt.figure(figsize=(21,7))
style = {'marker': 'o', 'markersize': 10, 'alpha': 0.8, 
         'linestyle': '--', 'linewidth': 3}

# Representación gráfica de los valores de la prueba F
plt.subplot(121)
plt.plot(dsup[1:], pval[1:], **style)
plt.ylabel('F-test probability' , fontsize=20);
plt.xlabel('Grado de polinomio', fontsize=20)
plt.xlim(3.5, 11.5)
plt.ylim(-0.03, 1.1)

# PuttiRepresentación residuos
plt.subplot(122)
style = {'marker': 'o', 'markersize': 8,
         'linestyle': '--', 'linewidth': 3}
for d in np.arange(3, degree_max):
    par, cov = fit_polynom(d)
    RSScum = np.cumsum(np.abs(y-pol_func(x, *par)))
    plt.plot(x, RSScum, label='pol{}'.format(d), **style)
plt.legend(frameon=False, fontsize='xx-large');
plt.ylabel('Cumulative Sum of $(y-f(x))^2$', fontsize=20);
```

Podemos observar en estos dos gráficos que, efectivamente, comparar polinomios de bajo grado (como $d=4$ y $d=3$) resulta en una diferencia q grande (y, por lo tanto, una probabilidad baja en la prueba F), mientras que comparar polinomios de alto grado (p. ej., $d=6$ y $d=5$) resulta en una predicción muy similar, lo que implica una probabilidad considerable en la prueba F. Se observa que la probabilidad de que $d=10$ sea similar a $d=10$ es bastante baja, lo que probablemente se pueda entender por un RSS bastante diferente (último punto en el gráfico de la derecha para `pol11` y `pol10`).
